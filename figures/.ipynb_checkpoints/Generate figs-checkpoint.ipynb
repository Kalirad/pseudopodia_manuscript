```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib.path as mpath
import matplotlib.lines as mlines
import matplotlib.patches as mpatches
from matplotlib.collections import PatchCollection
import matplotlib.patches as patches
```

```python
x = np.arange(0, 10, 1)
fig = plt.figure(figsize=(10,10))
plt.rcParams.update({'font.family': 'Arial'})
plt.rcParams.update({'font.size': 40})
ax = fig.add_subplot(111)
ax.spines['right'].set_visible(True)
ax.spines['top'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.spines['left'].set_linewidth(3)
ax.spines['bottom'].set_linewidth(3)
ax.spines['right'].set_linewidth(3)
ax.spines['top'].set_linewidth(3)
ax.spines['bottom'].set_color('black')
ax.spines['left'].set_color('black')
#ax.spines['left'].set_position(('outward', 15))
#ax.spines['bottom'].set_position(('outward', 15))

circ = patches.Circle((0., .5), 0.2, alpha=1, fc='dodgerblue')
ax.add_patch(circ)

circ = patches.Circle((0.5, 0.), 0.2, alpha=1, fc='red')
ax.add_patch(circ)

#circ = patches.Circle((0., 0.5), 0.1, alpha=1, fc='lightblue')
#ax.add_patch(circ)

rect = patches.Rectangle((0.45, 0.48), 0.27, 0.1, alpha=1, fc='black')
ax.add_patch(rect)

ar = patches.Arrow(0.44, 0.53,-0.13, 0., alpha=1, fc='black', width=0.05)
ax.add_patch(ar)

ar = patches.Arrow(0.5, 0.47,0, -0.13, alpha=1, fc='black', width=0.05)
ax.add_patch(ar)

ar = patches.Arrow(0.56, 0.725,0, -0.13, alpha=1, fc='black', width=0.05)
ax.add_patch(ar)


plt.plot(x,x, lw=4, color='silver', zorder=0, ls ='--')


plt.tick_params(axis='x', which='both', bottom='on', top='off',
    labelbottom='on', size=20, width=3, direction='out', pad=25, colors='black')
plt.tick_params(axis='y', which='both', left='on', right='off',
    labelbottom='on', size=20, width=3, direction='out', pad=25, colors='black')
#plt.ylabel(r'Simple DMIs', fontsize= 50, labelpad=20)
plt.xlabel(r'Phenotype B', fontsize= 50, labelpad=20)
plt.ylabel(r'Phenotype A', fontsize= 50, labelpad=20)

plt.xticks([])
plt.yticks([])

ax.xaxis.label.set_color('black')

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles,labels, loc='best', fontsize='medium', \
          handlelength=0.5, handletextpad=0.4, framealpha=0, labelspacing=0.1,\
          numpoints=1, markerscale=0.3)
plt.text(0.4525,0.5, r'Switch', fontsize= 50,  color='white')
plt.text(0.48,0.75, r'Noise', fontsize= 50, color='black')
plt.text(0.36,0.55, r'$P_{A}$', fontsize= 30, style='italic', color='black')
plt.text(0.51,0.4, r'$P_{B}$', fontsize= 30, style='italic', color='black')
plt.ylim(0, 1);
plt.xlim(0, 1);
#plt.show()
plt.savefig('fig1.eps', bbox_inches='tight', dpi=300)
```
